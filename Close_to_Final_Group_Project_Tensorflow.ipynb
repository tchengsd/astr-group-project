```python
import numpy as np
```

```python
#change the file name to correspond to your text file location
data = np.genfromtxt('./ASTR19_F24_group_project_data.txt', dtype=[('day', 'i8'), ('time', 'U6'), ('height', 'f8')])

#empty lists to be filled with column data
day = []
time = []
height = []

for i in range(82):
    #assign first, second and third columns to day, time, and height
    day.append(data[i][0])
    time.append(data[i][1])
    height.append(data[i][2])

print(day)
print('')
print(time)
print('')
print(height)
```

```python
percents = []
for times in time:
    hourmin = times.split(':')
    percent = (float(hourmin[0]) + float(hourmin[1])/60) /24
    percents.append(percent)
daytime = [percent + daynum for percent, daynum in zip(percents, day)]
```

```python
daytime = np.array(daytime, dtype=np.float32)
height = np.array(height, dtype=np.float32)

# Amplitudes
a_est = ((max(height) - min(height)) / 2)
a_est = tf.Variable(a_est, dtype=tf.float32)

d_est = a_est / 2
d_est = tf.Variable(d_est, dtype=tf.float32)

# Periods

b_est = 2*np.pi
b_est = tf.Variable(b_est, dtype=tf.float32)

e_est = np.pi/15
e_est = tf.Variable(e_est, dtype=tf.float32)

# Phases

c_est = 0
c_est = tf.Variable(c_est, dtype=tf.float32)

f_est = 2
f_est = tf.Variable(f_est, dtype=tf.float32)

# Offset

g_est = np.mean(height)
g_est = tf.Variable(g_est, dtype=tf.float32)

@tf.function
def f_tide(x):
    """A more complex tidal model function."""
    return (a_est * tf.math.sin(b_est * x + c_est)) + (d_est * tf.math.sin(e_est * x + f_est)) + g_est

@tf.function
def loss_function(y_true, y_pred):
    return tf.keras.losses.MeanSquaredError()(y_true, y_pred)


# Define the optimizer (Adam optimizer)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Training loop
epochs = 5000
display_step = 1000

for epoch in range(epochs):
    if ((epoch % display_step) == 0):
        pred = f_tide(daytime)
        loss = loss_function(pred, height)
        print(f"Epoch {epoch} | Loss = {loss.numpy()}, A: {a_est.numpy()}, B: {b_est.numpy()}, C: {c_est.numpy()}, D: {d_est.numpy()}, E: {e_est.numpy()}, F: {f_est.numpy()}, G: {g_est.numpy()}")

    with tf.GradientTape() as tape:
        pred = f_tide(daytime)
        loss = loss_function(pred, height)

    gradients = tape.gradient(loss, [a_est, b_est, c_est, d_est, e_est, f_est, g_est])
    optimizer.apply_gradients(zip(gradients, [a_est, b_est, c_est, d_est, e_est, f_est, g_est]))


print(f"Epoch {epoch} | Loss = {loss.numpy()}, A: {a_est.numpy()}, B: {b_est.numpy()}, C: {c_est.numpy()}, D: {d_est.numpy()}, E: {e_est.numpy()}, F: {f_est.numpy()}, G: {g_est.numpy()}")
print('Done!')
```

```python
def sinusoid(x, A, B, C, D, E, F, G):
    return (A * np.sin(B * x + C)) + (D * np.sin(E * x + F)) + G

x = np.asarray(np.random.uniform(low=min(daytime), high=max(daytime), size=100), dtype=np.float32)
y = np.asarray(sinusoid(x, a_est, b_est, c_est, d_est, e_est, f_est, g_est) + 0.25*np.random.randn(len(x)), dtype=np.float32)
y_err = np.full(100, 0.25, dtype=np.float32)

idx_model = np.argsort(x)
y_data = y[idx_model]
y_model = sinusoid(x[idx_model], a_est, b_est, c_est, d_est, e_est, f_est, g_est)

residuals = y_data - y_model
residuals = residuals.numpy()

print(residuals)

f, ax = plt.subplots(1, 1, figsize=(7, 7))

for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize(14)
for tick in ax.yaxis.get_ticklabels():
    tick.set_fontsize(14)


ax.errorbar(x, residuals, y_err, fmt='o', label='Residuals')
ax.set_ylim([-2, 2])
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
plt.legend(frameon=False, fontsize=14, handletextpad=0)
plt.show()
```

```python
residual_mean = np.mean(residuals)
residual_std = np.std(residuals)

def gaussian(x, mu, s):
    return 1./(2.*np.pi*s**2)**0.5 * np.exp(-0.5*((x-mu)/s)**2)

f, ax = plt.subplots(1, 1, figsize=(7, 7))
for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize(14)
for tick in ax.yaxis.get_ticklabels():
    tick.set_fontsize(14)

ax.hist(residuals, bins=10, range=(-0.75,0.75), alpha=0.5, edgecolor='white', density=True)

x_g = np.linspace(-5*residual_std, 5*residual_std, 1000)
ax.plot(x_g, gaussian(x_g, residual_mean, residual_std), color='red')
ax.set_xlim([-2, 2])
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
plt.show()
```
