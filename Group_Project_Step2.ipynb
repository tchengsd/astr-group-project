1) Download this data setLinks to an external site. that lists some tidal measurements Santa Cruz in January and February 2022. In your Jupyter notebook, define an oscillatory function in Python to describe the intra- and inter-day variations in the tide.

```python
import numpy as np
```

```python
#change the file name to correspond to your text file location
data = np.genfromtxt('./ASTR19_F24_group_project_data.txt', dtype=[('day', 'i8'), ('time', 'U6'), ('height', 'f8')])

#empty lists to be filled with column data
day = []
time = []
height = []

for i in range(82):
    #assign first, second and third columns to day, time, and height
    day.append(data[i][0])
    time.append(data[i][1])
    height.append(data[i][2])

print(day)
print('')
print(time)
print('')
print(height)
```

```python
percents = []
for times in time:
    hourmin = times.split(':')
    percent = (float(hourmin[0]) + float(hourmin[1])/60) /24
    percents.append(percent)
daytime = [percent + daynum for percent, daynum in zip(percents, day)]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(daytime, height, '-o')
plt.xlabel('Day')
plt.ylabel('Height of tide (ft)')
plt.show()
```

```python
from scipy import optimize
```

```python
def f_tide(x, a, b, c, d, e, f, g):
    """A more complex tidal model function."""
    return (a * np.sin(b * x + c)) + (d * np.sin(e * x + f)) + g

# Experiment with different initial parameters
# 1. Amplitude:

a_est = ((max(height) - min(height)) / 2)

d_est = a_est / 2

# 2. Period:

b_est = 2*np.pi

e_est = np.pi/15

# 3. Phase:
# Rough estimate based on visual inspection - adjust if needed
c_est = 0
f_est = 0
g_est = np.mean(height)

print(a_est, b_est, c_est, d_est, e_est, f_est, g_est)

# Curve Fitting

initial_params = [a_est, b_est, c_est, d_est, e_est, f_est, g_est]

# Perform curve fitting
params, params_cov = optimize.curve_fit(
    f_tide, daytime, height, sigma=np.full(len(daytime), 0.25),
    p0=initial_params
)

# Extract fitted parameters
a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit = params

# Generate fitted values
height_fit = f_tide(np.array(daytime), a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit)

print(height_fit)
```

```python
plt.plot(daytime, height_fit, label="Fitted Model")
plt.plot(daytime, height, '-o', label="Observed Data", alpha=0.5)
plt.xlabel('Day')
plt.ylabel('Height of tide (ft)')
plt.legend()
plt.show()
```

```python
residuals = []
i = 0
while i < len(height):
    #subtracted the best fit from the data to get the residuals
    residuals.append(height[i] - height_fit[i])
    i += 1

print(residuals)
print(min(residuals), max(residuals))
```
